```python
from scipy.optimize import linprog
import matplotlib.pyplot as plt
import numpy as np

# Función que se desea maximizar
c = [-10, -15]  # Usamos negativos porque linprog trata de minimizar

# Matriz de coeficientes de las restricciones
A = [[2, 4], [3, 2]] # Horas de mano de obra y Unidades de materia prima

# Vector con los valores de las desigualdades
b = [100, 120]

# Restricciones de no-negatividad
x0_bounds = (0, None)
x1_bounds = (0, None)

res = linprog(c, A_ub=A, b_ub=b, bounds=[x0_bounds, x1_bounds], method='simplex')

print(f'Producto A: {res.x[0]} unidades')
print(f'Producto B: {res.x[1]} unidades')
print(f'Ganancia máxima: {-res.fun} dólares')  

x = np.linspace(0, 60, 200)
y1 = (100 - 2*x) / 4
y2 = (120 - 3*x) / 2

plt.figure(figsize=(10, 10))
plt.plot(x, y1, label=r'$2x + 4y \leq 100$')
plt.plot(x, y2, label=r'$3x + 2y \leq 120$')
plt.xlim((0, 60))
plt.ylim((0, 40))
plt.axhline(0, color='black',linewidth=0.5)
plt.axvline(0, color='black',linewidth=0.5)
plt.fill_between(x, np.minimum(y1, y2), where=(x<=50), color='gray', alpha=0.5)
plt.scatter(res.x[0], res.x[1], color='red')
plt.xlabel('Producto A')
plt.ylabel('Producto B')
plt.legend()
plt.grid(True)
plt.show()
```

```python
from scipy.optimize import linprog

# Función a Maximizar
c = [-1500, -2500, -4000]  # Usamos negativos porque linprog trata de minimizar

# Matriz de coeficientes de las restricciones
A = [[2, 3, 5], [1, 2, 3]]

# Vector con los valores a la derecha de las desigualdades
b = [15, 10] # Horas de Trabajo y Personas

# Restricciones de no-negatividad
x0_bounds = (0, None)
x1_bounds = (0, None)
x2_bounds = (0, None)

res = linprog(c, A_ub=A, b_ub=b, bounds=[x0_bounds, x1_bounds, x2_bounds], method='simplex')

print(f'Tarea A: {res.x[0]} unidades')
print(f'Tarea B: {res.x[1]} unidades')
print(f'Tarea C: {res.x[2]} unidades')
print(f'Ganancia máxima: {-res.fun} dólares')

```
